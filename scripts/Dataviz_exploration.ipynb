# Image Synthesis 

It refers to the process of creating realistic images from random input data. This is usually done using generative adverserial networks (GANs) which was introduced in 2014 by [Ian Goodfellow et al](http://papers.nips.cc/paper/5423-generative-adversarial-nets). GANs usually work by having two networks the first network is called a descriminator which can differetniaate between fake and real images. The second network is called a generator which given random data it can generate realistic images from the learnt distribution of the training data. GANs have improved since 2014. 

![alt text](https://pbs.twimg.com/media/Deex0DoU0AABQK_.jpg)

One common problem when training on a huge set of training data like ImageNet for instance is the statilibity of the training process. When the training process is not stable this causes the images not to look as realisitic as we wish. Usually we use the inception score **IS** which is a metric to evaluate the the quality of synthesized images images and their diversity.  Given $x \sim p_g$ sampled from the probability distribution of the generated images then 

$$\mathrm{IS}(G) = \mathrm{exp}\left(\mathbb{E}_{x\sim p_g} D_{KL}  (p(y|x) || p(y))\right)$$

an Inception v3 model is used to evaluate the statistics of the generated samples. $KL$ is the KL-divergence is a metric to evaluate the "closeness" of two distributions. Another metric is Fr√©chet Inception Distance **FID**. Given a realistic image $x$ and a generated image $g$ we calcuate the score as 

$$\mathrm{FID} (x,g) = \Vert \mu_x - \mu_g\Vert_2^2 + \mathrm{Tr}\left( \Sigma_x+\Sigma_g  - 2 \sqrt{\Sigma_x \Sigma_g}\right) $$

where $\mathrm{Tr}$ is the trace of a matrix and $\Sigma$ is the covariance matrix. 

# BigGan

"Large Scale GAN Training for High Fidelity Natural Image Synthesis" by [Andrew Brock et al.](https://arxiv.org/abs/1809.11096) from DeepMind  was introdcued just at the end of Septemeter this year. It provides a state of the art results of image synthesis. It provides IS of 166.3 and   FID of 9.6, improving over the previous best IS of 52.52 and FID of 18.65 which is a huge improvment over the previous results. The results are impressive as we see from the following images as it seems difficult to imagine that these images were computer generated!

![alt text](https://cdn-images-1.medium.com/max/2000/1*Yw2KxjmIkj8yqS-ykLCQCQ.png)

BigGans are a scaled up version of previous methods by providing larger networks and larger batch size. According to the paper 

" *GANs benefit dramatically from scaling, and train models with two to four times as many parameters and eight times the batch size compared to prior art"*

The models are  trained on 128 to 512 cores of a Google TPU which is pretty huge.  BigGan is a conditional GAN where class indices are  inserted into an embedding layer in order to train the generator network to condition on the class index. Moreover, the authors used 
 a variant of hierarchical latent spaces, where the noise vector z is inserted into multiple layers
of G at different depths which allows the latent vector to act on features on different levels.

Store the number of the labels and the size of the latent sapce $z$ to sample from 

```python
vocab_size = inputs['y'].get_shape().as_list()[1]
latent_size = inputs['z'].get_shape().as_list()[1]

print('Number of labels ', vocab_size)
print('The size of the latent space ', latent_size)
```

Note that the model takes label inputs as one hot encoded which maps each class $c \in [0, 1000)$ to a vector of size $1000$ with all zeros except the index of the corrosponding class

```python
#helper functions to generate the encoding
def one_hot(index, vocab_size=1000):
  index = np.asarray(index)
  if len(index.shape) == 0:
    index = np.asarray([index])
  assert len(index.shape) == 1
  num = index.shape[0]
  output = np.zeros((num, vocab_size), dtype=np.float32)
  output[np.arange(num), index] = 1
  return output
```

# Truncation Trick 

Previous work on GANs samples the latent vector $z \sim \mathcal{N}(0, I)$ as a normal distirubtion with the identity convariance matrix. OTOH, the authors of BigGans used a truncated normal distriubtion in a certain region $[-a, a]$ for $a \in \mathbb{R}^+$where the sampled values outside that region are resampled to fall in the region. This resulted in better results of both **IS** and **FID** scores. The drawback of this is a reduction in the overall variety of vector sampling. Hence there is a trade-off between sample quality and variety ofr a given network **`G`**. 

Here we set the default truncation threshold to be $2$ i.e  $z$ values are sampled from the  region $[-2, 2]$. Note that the optional variable `seed` takes integer values and it's used to reserve the state of randomness for resampling. Hence, if you use the same seed you will get the same $z$ vector .

```python
def get_zy(index, trunc = 1., batch_size = 1, seed = None):
  #convert the label to one hot encoding 
  y = one_hot(index)
  
  #sample a batch of z-vectors 
  z = truncnorm.rvs(-2, 2, size=(batch_size, latent_size), random_state = np.random.RandomState(seed)) * trunc
  return z, y
```

Note that with normal distribution values near zero are chosen with higher probability 

```python
z, _ = get_zy(np.random.choice(range(0, vocab_size), size = 1),  seed = 2)

y = np.linspace(0, latent_size, latent_size)

plt.scatter(y, z)
plt.show()
```

# Latent Space Interpolation

Given two vectors $v_1, v_2$ and an integer $N$  we want to interpolate between them i.e create a mapping that moves between two vectors with $N$ vectors. 

## Linear Interpolation 

It is the simplest types of interpolation where we use the simple function 

$$F(v_1, v_2, N) = x v_2 + (1-x) v_1 \,\, ,\,\,\, x \in \left(0,\frac{1}{N}, \cdots, \frac{N}{N}\right)$$

```python
def interpolate_linear(v1, v2, num_steps):
  vectors = []
  for x in np.linspace(0.0, 1.0, num_steps):
    vectors.append(v2*x+v1*(1-x))  
  return np.array(vectors)
```

## Hypersphere Interpolation

For each iteration $i \in (0, N-1)$

$$ z = v_1 + \left( v_2 \frac{\Vert v_1 \Vert }{\Vert v_2 \Vert}  - v_1\right) \times \frac{i}{N-1}$$

Then update $z$ as 

$$\hat{z} =  \Vert v_1 \Vert \times \frac{z}{\Vert z \Vert}$$

```python
def interpolate_hypersphere(v1, v2, num_steps):

  v1_norm = LA.norm(v1)
  v2_norm = LA.norm(v2)
  v2_normalized = v2 * (v1_norm / v2_norm)

  vectors = []
  for step in range(num_steps):
    interpolated = v1 + (v2_normalized - v1) * step / (num_steps - 1)
    interpolated_norm =  LA.norm(interpolated)
    interpolated_normalized = interpolated * (v1_norm / interpolated_norm)
    vectors.append(interpolated_normalized)
  return np.array(vectors)
```
